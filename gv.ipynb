# Drawing with Graphviz

```python
import itertools

import graphviz
```

```python
g0 = graphviz.Graph()  # Uses the 'dot' engine by default.
g0.edge('A', 'B')
g0.edge('A', 'C')
g0.edge('B', 'C')
g0.edge('C', 'D')
g0.node('E')
g0
```

```python
g0.engine = 'circo'  # Sometimes circo is better for undirected graphs.
g0
```

```python
g1 = graphviz.Digraph()
g1.edge('A', 'B')
g1.edge('A', 'C')
g1.edge('B', 'D')
g1.edge('B', 'E')
g1.edge('C', 'F')
g1
```

```python
g2 = graphviz.Digraph()
g2_nulls = itertools.count()

def make_null() -> str:
    """Make a point node in g2 distinct from any other. Return its name."""
    name = f'null{next(g2_nulls)}'
    g2.node(name, shape='point')
    return name

g2.edge('A', 'B')
g2.edge('A', 'C')
g2.edge('B', 'D')
g2.edge('B', 'E')
g2.edge('C', make_null())
g2.edge('C', 'F')
g2
```

```python
# Add the remaining nulls for consistency.
for node in 'D', 'E', 'F':
    for _ in range(2):  # Left and right.
        g2.edge(node, make_null())
g2
```

```python
g2.graph_attr['rankdir'] = 'LR'  # Left to right. (TB, BT, LR, and RL all work.)
g2
```

```python
g3 = graphviz.Digraph()

g3.edge('A','B')
g3.edge('B','C')
g3.edge('C','D')
g3.edge('D','E')
g3.edge('E','A')
g3.edge('A','D')
g3.edge('D','B')
g3.edge('B','E')
g3.edge('E','C')
g3.edge('C','A')
g3
```

```python
g3.engine = 'circo'
g3
```

```python
g3.engine = 'fdp'
g3
```

```python
g4 = graphviz.Digraph()

g4.edge('A','B', color='purple')
g4.edge('A','C')
g4.edge('D','C', color='red')
g4.edge('D','C', color='blue')
g4.node('E', color='green')
g4.node('F', color='gold')
g4.edge('F','A', color='yellow')

g4
```
